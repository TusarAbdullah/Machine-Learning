```python
import gradio as gr
```

```python
def rent_recom(x):
    rent = reg3.predict([[x]])
    return rent
```

```python
rent_recom(3000)
```

```python
interface = gr.Interface(
    fn = rent_recom,
    inputs = ['number'],
    outputs = ['number']
).launch(share=False)     #False dity hoi nije korar jonno... true public er jonno
```

```python
interface = gr.Interface(
    fn = rent_recom,
    inputs = ['number'],
    outputs = ['number']
).launch(share=True)     #False dity hoi nije korar jonno... true public er jonno
```
